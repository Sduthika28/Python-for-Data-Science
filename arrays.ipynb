```python
# Assuming datacopy is your DataFrame
x = datacopy[["area", "bedrooms", "bathrooms", "stories", "parking"]]
y = datacopy["price"]

# Specify feature names explicitly
feature_names = ["area", "bedrooms", "bathrooms", "stories", "parking"]
x.columns = feature_names

# Create and fit the linear regression model
lrg = linear_model.LinearRegression()
lrg.fit(x, y)

# Make predictions
predicted = lrg.predict([[2400, 3, 1, 1, 0]])
```
