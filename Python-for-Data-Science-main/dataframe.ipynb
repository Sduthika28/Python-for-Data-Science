```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
os.chdir("C:\\Users\\SADHANA\\OneDrive\\Documents\\Desktop\\PYTHON\\Jupyter Notebook\\CSV")
data=pd.read_csv("datavisdf.csv")
print(data)
```

```python
#Scatter Plot
plt.scatter(data["Age"],data["Salary"],c="green")
plt.title("Scatter Plot")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.show()
```

```python
#HISTOGRAM
plt.hist(data["Age"],color="pink",edgecolor="green",bins=7)
plt.xlabel("Age")
plt.ylabel("Salary")
plt.show()
```
